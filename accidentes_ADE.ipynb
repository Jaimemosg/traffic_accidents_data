# Anonymus Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from os import listdir
import pandas as pd
pd.options.display.max_columns = None
```

# Open data

```python
files_list =listdir()
files_list
```

```python
Accidentes = pd.DataFrame() 

for file in files_list:
    if ('.csv' in file):
        temp_df = pd.read_csv(file, sep=',')
#         print(len(temp_df.columns))
        Accidentes = Accidentes.append(temp_df)
Accidentes.head()        
```

```python
sns.countplot(data = Accidentes, x = 'CLASE');
plt.xticks(rotation=90);
```

```python
print(Accidentes.columns)
print('Number of columns:', len(Accidentes.columns))
print('Number of rows:', len(Accidentes))
```

```python
categoricals = ['PERIODO', 'CLASE', 'GRAVEDAD','DIA_NOMBRE', 'DISENO','MES']

fig = plt.figure(figsize=(15, 20))

for i, cat in enumerate(categoricals):
    plt.subplot(4, 3, i+1)
    sns.countplot(data = Accidentes, x = cat);
    xtitle = ' '.join(cat.title().split('_'))
    plt.xlabel(xtitle)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.6)
```
